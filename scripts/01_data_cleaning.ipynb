```r
library(tidyverse)

df <- read_csv("../data/telecom_churn.csv") 

glimpse(df)
```

```r
# Load raw data
raw_df <- read_csv("../data/telecom_churn.csv")

# Quick count of missing values
sum(is.na(raw_df))
```

```r
df_clean <- raw_df %>%
  # 1. Drop customerID (not useful for prediction)
  select(-customerID) %>%
  
  # 2. Convert SeniorCitizen from 0/1 to "No"/"Yes" for clarity, then to factor
  mutate(SeniorCitizen = ifelse(SeniorCitizen == 1, "Yes", "No")) %>%
  
  # 3. Convert all Character columns to Factors (required for ML in R)
  mutate(across(where(is.character), as.factor)) %>%
  
  # 4. Drop rows with missing TotalCharges (Usually ~11 rows where tenure is 0)
  drop_na()

# Verify the cleanup
glimpse(df_clean)
```

```r
# Check the target variable distribution
df_clean %>%
  count(Churn) %>%
  mutate(prop = n / sum(n))
```

```r
# Save to data folder
write_rds(df_clean, "../data/telecom_churn_clean.rds")

print("Data cleaning complete. Clean dataset saved to data/telecom_churn_clean.rds")
```
